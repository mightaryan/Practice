```python
# Unsupervised Learning: When we have unlabeled data and want to find hidden patterns or groupings within the data.
# Kmeans Clustering : When we want to partition data into distinct groups (clusters) based on feature similarity. 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.cluster as cluster

# Example data: 2D points
data = {
    'Customer' : ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Age' : [25, 34, 22, 45, 33, 26, 40, 29],
    'Income' : [50000, 60000, 45000, 80000, 62000, 52000, 75000, 58000]
}

df = pd.DataFrame(data)

# K-means Clustering
"""Educlidean distance is used to measure the similarity between data points in K-means clustering.
It calculates the straight-line distance between two points in multi-dimensional space. """

"""Elbow Method is a technique used to determine the optimal number of clusters in K-means clustering.
It involves plotting the explained variance (or inertia) against the number of clusters and identifying
the 'elbow' point where the rate of decrease in inertia starts to slow down.   """

X = df[['Age', 'Income']]

model = cluster.KMeans(n_clusters=2, random_state=42, n_init=10)
df['Group'] = model.fit_predict(X)

# Plotting the clusters
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Income'], c=df['Group'], cmap='viridis', marker='o')
plt.title('K-means Clustering of Customers')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
print(df)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np


data = {
    'Age': [25, 34, 22, 45, 33, 26, 40, 29],
    'Income': [50000, 60000, 45000, 80000, 62000, 52000, 75000, 58000],
    'SpendingScore': [39, 81, 6, 77, 40, 76, 94, 3],
    'Saving': [2000, 3000, 1500, 4000, 3200, 2100, 3800, 2900]
}

df = pd.DataFrame(data)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

print("Scaled Data:")
print(scaled_data)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

df_pca = pd.DataFrame(data=pca_result, columns=['Principal Component 1', 'Principal Component 2'])
explained_variance = pca.explained_variance_ratio_
print("\nExplained Variance Ratio:")
print(np.round(explained_variance * 100, 2))
print("\nPCA Result:")
print(df_pca)

plt.figure(figsize=(8, 6))
plt.scatter(df_pca['Principal Component 1'], df_pca['Principal Component 2'])
plt.title('PCA Result')
plt.grid()
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

```
