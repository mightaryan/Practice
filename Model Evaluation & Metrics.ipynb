```python
# accuracy = correct_predictions / total_predictions
# sklearn.metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


# Example usage
y_true = [0, 1, 1, 0, 1, 1, 0]
y_pred = [0, 1, 0, 0, 1, 1, 1]
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")    
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix

# Example usage
y_true = [0, 1, 1, 0, 1, 1, 0]
y_pred = [0, 1, 0, 0, 1, 1, 1]
cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
# Mean Absolute Error (MAE) and Mean Squared Error (MSE) and Root Mean Squared Error (RMSE)
# Mean Absolute Error (MAE) = (1/n) * Σ |y_true - y_pred|
# Mean Squared Error (MSE) = (1/n) * Σ (y_true - y_pred)²
# Root Mean Squared Error (RMSE) = sqrt(MSE)
from sklearn.metrics import mean_absolute_error, mean_squared_error
from math import sqrt

# Example usage
students_marks_true = [85, 90, 78, 92, 88, 76, 95, 89]
students_marks_pred = [80, 88, 79, 91, 87, 75, 94, 90]
mae = mean_absolute_error(students_marks_true, students_marks_pred)
mse = mean_squared_error(students_marks_true, students_marks_pred)
rmse = sqrt(mse)
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")


```
